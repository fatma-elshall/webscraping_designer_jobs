**import libraries**

```python
from bs4 import BeautifulSoup as bs  #for scrapping
```

```python
from urllib.request import urlopen  #for url of page
```

**HTML of page**

```python
url='https://wuzzuf.net/search/jobs/?a=spbg&filters%5Bcareer_level%5D%5B0%5D=Entry%20Level&q=graphic%20designer'
```

**create client based request to get url**

```python
client=urlopen(url)
```





*   now i have a connection with page






**reading code of page **

```python
html=client.read()
html
```

```python
client.close()
```

**create html parser using beautifulsoup**



*   to handle the code and scrapping information



```python
soup=bs(html,'html.parser')
soup
```



*   create a container for needed data



```python
containers=soup.find_all('div',{'class':'css-1gatmva e1v1l3u10'})
len(containers)
```

```python
bs.prettify(containers[0])
```



*   now lets scrapping page elemnts



```python
#jop title
jtitle=containers[0].find_all('h2',{'class':'css-m604qf'})
jtitle[0].text
```

```python
#company name
cname=containers[0].find_all('a',{'class':'css-17s97q8'})
cname[0].text
```

```python
#place
jplace=containers[0].find_all('span',{'class':'css-5wys0k'})
print(jplace[0].text)
```

```python
#type
jtype=containers[0].find_all('a',{'class':'css-n2jc4m'})
jtype[0].text
```

```python
#experiance
jexperiance=containers[0].find_all('a',{'class':'css-o171kl'})
jexperiance[1].text
```



*   make a file to save data



```python
file=open('graphic designer jobs.csv','w')
header='job_title,job_company,job_place,job_type,Jexperiance\n'
file.write(header)
```

```python

# make for loop for all these
for i in containers:
  jtitle=i.find_all('h2',{'class':'css-m604qf'})  #title
  title=jtitle[0].text.strip()

  cname=i.find_all('a',{'class':'css-17s97q8'})  #company
  company=cname[0].text.strip()

  jplace=i.find_all('span',{'class':'css-5wys0k'})  #place
  place=jplace[0].text.strip()

  jtype=i.find_all('a',{'class':'css-n2jc4m'})  # type
  type1=jtype[0].text.strip()

  jexperiance=i.find_all('a',{'class':'css-o171kl'})
  experiance=jexperiance[1].text.strip()

  #print(title+','+ company+',' + place+','+ type1+','+ experiance+'\n')
  file.write(title+','+ company+',' + place+','+ type1+','+ experiance+'\n')
file.close()







```
